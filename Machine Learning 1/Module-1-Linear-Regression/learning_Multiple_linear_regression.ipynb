## Step: 2

- encoding:
    - converting binry to 1/0
    - other categorical variables to dummy variables
- splitting train-test sets
- rescalling

```python
# yes/no variables
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
housing[varlist].head()
```

```python
housing[varlist] = housing[varlist].apply(lambda x: x.map({'yes':1, 'no':0}))
housing[varlist]
```

```python
housing.head()
```

```python
# creating dummy variables for furnishingstatus
# dropping redundant dummy variable

status = pd.get_dummies(housing.furnishingstatus, drop_first=True)
status.head()
```

```python
housing = pd.concat([housing,status],axis=1)
housing.head()
```

```python
housing.drop('furnishingstatus', axis=1,inplace=True)
```

```python
housing.head()
```

#### splitting train-test datasets

```python
df_train, df_test = train_test_split(housing, train_size=0.70, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

#### rescalling
1. Min-Max scaling (normalization ): betwen 0 and 1 -- (x - xmin)/(xmax - xmin)
2. Standardization: mean=0, sigma=1 -- (x-mu)/(sigma)

```python
# 1. Instantiate the object
scaler = MinMaxScaler()

# create a list of numerical variables 
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

# 2. fit on the data
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
df_train[num_vars].describe()
```

## Step: 3

### Training the model

```python
# heatmap
plt.figure(figsize=(12,10))
sns.heatmap(df_train.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
# X_train, y_train
X_train = df_train
y_train = df_train.pop('price')
```

```python
# add a constant
X_train_sm = sm.add_constant(X_train['area'])
```

```python
# build model
lr = sm.OLS(y_train, X_train_sm)

# fit
lr_model = lr.fit()
```

```python
# params
lr_model.params
```

```python
# summary
lr_model.summary()
```

```python
# add another variable
X_train_sm = X_train[['area','bathrooms']]

# add a constant
X_train_sm = sm.add_constant(X_train_sm)

# build model
lr = sm.OLS(y_train, X_train_sm)

# fit
lr_model = lr.fit()
```

```python
# summary
lr_model.summary()
```

```python
# add bedrooms
X_train_sm = X_train[['area','bathrooms','bedrooms']]

# add a constant
X_train_sm = sm.add_constant(X_train_sm)

# build model
lr = sm.OLS(y_train, X_train_sm)

# fit
lr_model = lr.fit()
```

```python
# summary
lr_model.summary()
```

### VIF

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
X = X_train.drop('semi-furnished', axis=1)
```

```python
# create another model

# add a constant
X_train_sm = sm.add_constant(X)

# build model
lr = sm.OLS(y_train, X_train_sm)

# fit
lr_model = lr.fit()

lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
X = X.drop('bedrooms', axis=1)
```

```python
# create another model

# add a constant
X_train_sm = sm.add_constant(X)

# build model
lr = sm.OLS(y_train, X_train_sm)

# fit
lr_model = lr.fit()

lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```
